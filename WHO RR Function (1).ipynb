```python
import numpy as np
import time
```

```python
def life_expectancy():
    #Call the function like this, no inputs required, numpy and time packages are required
    
    while True:
    #This loop checks that you specifically answer Y or N to consent question
        question=input("Do you consent to using advanced population data, which may include protected information, for better accuracy? (Y/N)").lower().strip()
        #Allows wide variety of valid input types, e.g. capital or lowercase and spcaes before or after the input
        
        if question=='y':
            while True: 
            #Each of these loops check that the variable is specified in the correct manner
                region=input("What is your Region? (Options: Rest of Europe, Middle East, European Union, Central America and Caribbean, North America, Asia, Oceania, Africa, South America) ")\
                .lower().strip()
                
                if region not in ('rest of europe','middle east','european union','central america and caribbean','north america','asia','oceania','africa','south america'):
                    print("Please input a valid region")
                    time.sleep(1)
                else:
                    break
                    
            while True:
                try:
                    infant_deaths=float(input("Number of Infant Deaths per 1000 population: "))
                    if infant_deaths>=0 and infant_deaths<=1000:
                        break
                    else:
                        #We require logical inputs into the model
                        print("Sorry, number must be between 0 and 1000")
                        time.sleep(1)
                except ValueError:
                    print("Please input valid number")
                    time.sleep(1)
                    
            while True:
                try:
                    under_5_deaths=float(input("Number of under-5 Deaths per 1000 population: "))
                    if under_5_deaths>=0 and under_5_deaths<=1000:
                        break
                    else:
                        print("Sorry, number must be between 0 and 1000")
                        time.sleep(1)
                except ValueError:
                    print("Please input valid number")
                    time.sleep(1)
                    
            while True:
                try:
                    adult_mortality=float(input("Adult Mortality Rates of both sexes (number dying between 15 and 60 years per 1000 population): "))
                    if adult_mortality>=0 and adult_mortality<=1000:
                        break
                    else:
                        print("Sorry, number must be between 0 and 1000")
                        time.sleep(1)
                except ValueError:
                    print("Please input valid number")
                    time.sleep(1)

            while True:
                try:
                    hiv=int(input("Deaths per 1 000 live births HIV/AIDS (0-4 years): "))
                    if hiv>=0 and hiv<=1000:
                        break
                    else:
                        print("Sorry, number must be between 0 and 1000")
                        time.sleep(1)
                except ValueError:
                    print("Please input valid number")
                    time.sleep(1)


            while True:
                try:
                    thin10=float(input("Prevalence of thinness among children and adolescents for Age 10 to 19 (%): "))
                    if thin10>=0 and thin10<=100:
                        break
                    else:
                        print("Sorry, number must be between 0 and 100")
                        time.sleep(1)
                except ValueError:
                    print("Please input valid number")
                    time.sleep(1)
                    
            while True:
                developing=input("Developing Status? (Y/N)").lower().strip()
                if developing not in ('y', 'n'):              
                    print("Please input Y or N")
                    time.sleep(1)
                else:
                    break
            
            #Turn into array ready to feed into model.
            data=[adult_mortality, infant_deaths, under_5_deaths, region=='central america and caribbean', developing=='y', region=='asia',\
                  hiv, thin10]
            
            #coefficients taken from trained model
            params=[
                    -0.04807027664173644,
                     -0.07413821883265603,
                     -0.03990297016813743,
                     1.5726542380682422,
                     -2.325774617472684,
                     0.4319855186573345,
                     0.11999984446632744,
                     -0.03103099478106397]
            constant=83.73010225642325
            
            #Compute the prediction manually (standard regression equation) and return
            prediction=np.dot(data,params)+constant
            years=int(prediction)
            days=round((prediction-years)*365)
            return print(f'Life expectancy prediction is: {years} years and {days} days')
            
        elif question=='n':
        #Equivalent for simplistic model
            while True: 
                region=input("What is your Region? (Options: Rest of Europe, Middle East, European Union, Central America and Caribbean, North America, Asia, Oceania, Africa, South America) ")\
                .lower().strip()
                
                if region not in ('rest of europe','middle east','european union','central america and caribbean','north america','asia','oceania','africa','south america'):
                    print("Please input a valid region")
                    time.sleep(1)
                else:
                    break
                    
            while True:
                try:
                    adult_mortality=float(input("Adult Mortality Rates of both sexes (number dying between 15 and 60 years per 1000 population): "))
                    if adult_mortality>=0 and adult_mortality<=1000:
                        break
                    else:
                        print("Sorry, number must be between 0 and 1000")
                        time.sleep(1)
                except ValueError:
                    print("Please input valid number")
                    time.sleep(1)
                    
            while True:
                developing=input("Developing Status? (Y/N)").lower().strip()
                if developing not in ('y', 'n'):              
                    print("Please input Y or N")
                    time.sleep(1)
                else:
                    break
                    
            data=[adult_mortality, region=='rest of central america and caribbean', region=='rest of europe', region=='oceania', developing=='y']
            params= [-0.07018342110543188,
                     2.8691800288136844,
                     2.119013890349195,
                     -0.8799595560940836,
                     -3.928152719275068]
            constant=85.04934865515236
            
            prediction=np.dot(data,params)+constant
            years=int(prediction)
            days=round((prediction-years)*365)
            return print(f'Life expectancy prediction is: {years} years and {days} days')
        else:
            print("Please answer Y or N")
            time .sleep(1)
```
