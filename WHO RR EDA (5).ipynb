### **6. Visualization & Correlation**

```python
## Displaying the heatmap for the correlation of the columns

plt.figure(figsize = (16, 8))
sns.heatmap(whodf.corr(numeric_only=True), 
            annot = True, 
            fmt = '.2', 
            cmap='rocket',
            mask = np.triu(whodf.corr(numeric_only=True)), 
            vmin=-1,
            vmax=1)

plt.title("Correlation of the features", weight='bold')
plt.show()
```

#### **Findings:**
###### Life_expectancy has strong correlation with: Infant_deaths, Under_five_deaths, Adult_mortality
###### Economy_status_Developed & Economy_status_Developing are complement sets

### **7. Analysing few columns**

```python
# Checking that Economy_status_Developing and Economy_status_Developed are complement of each other!

whodf['Country'][ whodf['Economy_status_Developing'].astype('bool') == whodf['Economy_status_Developed'].astype('bool')].count()
```

```python
# Creating log for GDP

whodf['GDP_per_capita_log'] =  np.log((whodf['GDP_per_capita']))
whodf.head()
```

```python
# Creating scatterplot to display the comparision between GDP log & Life expectancy

fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False)
sns.scatterplot(y = whodf['Life_expectancy'], x = whodf['GDP_per_capita'], ax = ax1)
sns.scatterplot(y = whodf['Life_expectancy'], x = np.log(whodf['GDP_per_capita_log']), ax = ax2)
fig.tight_layout()
fig.show()
```

#### **Findings:**
###### GDP per capita is exponential, so we have taken the log and it seems linear!
